# Lotka–Volterra equations
The Lotka–Volterra equations, also known as the predator–prey equations, are a pair of first-order nonlinear differential equations, frequently used to describe the dynamics of biological systems in which two species interact, one as a predator and the other as prey. The populations change through time according to the pair of equations:

$$\frac{dx}{dt} = \alpha x - \beta xy$$
$$\frac{dy}{dt} = \delta xy - \gamma y$$

where
- x is the number of prey (for example, rabbits);
- y is the number of some predator (for example, foxes);
- $\frac{dy}{dt}$ and $\frac{dx}{dt}$ represent the instantaneous growth rates of the two populations;
- t represents time;
- $\alpha, \beta, \gamma, \delta$ are positive real parameters describing the interaction of the two species.

```python
def var_prey(alpha, x , beta, y):
    return alpha * x - beta * x * y 
def var_pred(delta, x, gamma, y):
    return delta *  x * y - gamma * y
```

```python
import matplotlib.pyplot as plt
import numpy as npy
alpha = 1 
beta = 0.1
delta = 0.5
gamma = 0.02

nr_rabbits = 100
nr_foxes = 20

population_rabbits = []
population_foxes = []



step = 0.0001

t = npy.arange(0,40,step)

for i in t:
    population_rabbits.append(nr_rabbits)
    population_foxes.append(nr_foxes)
    dx_rabbits = nr_rabbits * (alpha - beta * nr_foxes) 
    dy_foxes = -nr_foxes * (delta - gamma * nr_rabbits) 
    
    nr_rabbits = nr_rabbits + dx_rabbits * step
    nr_foxes = nr_foxes + dy_foxes * step
    


print(population_foxes)
plt.plot(t,population_rabbits)
plt.plot(t,population_foxes)
plt.show()
```
